```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the TinyMNIST dataset using NumPy
def load_data():
    train_data = pd.read_csv('TinyMNIST/trainData.csv').values
    train_labels = pd.read_csv('TinyMNIST/trainLabels.csv').values.flatten()
    test_data = pd.read_csv('TinyMNIST/testData.csv').values
    test_labels = pd.read_csv('TinyMNIST/testLabels.csv').values.flatten()
    # Normalize pixel values to [0, 1]
    X_train, X_test = train_data / 255.0, test_data / 255.0

    return X_train, train_labels.astype(int), X_test, test_labels.astype(int)

# Activation functions
def relu(x):
    return np.maximum(0, x)

def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)

# Derivatives
def relu_derivative(x):
    return (x > 0).astype(float)

def one_hot_encode(y, num_classes):
    return np.eye(num_classes)[y]

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

# Initialize weights and biases
def initialize_weights(sizes):
    weights = [np.random.randn(sizes[i], sizes[i+1]) * np.sqrt(2.0/sizes[i]) for i in range(len(sizes)-1)]
    biases = [np.zeros((1, sizes[i+1])) for i in range(len(sizes)-1)]
    return weights, biases

def forward_propagation(X, weights, biases, activations):
    A = X
    caches = []
    for i in range(len(weights)):
        Z = np.dot(A, weights[i]) + biases[i]
        A = activations[i](Z)
        caches.append((A, Z))
    return A, caches

# Backpropagation
def back_propagation(X, y, weights, caches, activations_derivative):
    grads_w = []
    grads_b = []
    A, Z = caches[-1]
    m = X.shape[0]
    dZ = A - y
    dW = np.dot(caches[-2][0].T, dZ) / m
    db = np.sum(dZ, axis=0, keepdims=True) / m
    grads_w.append(dW)
    grads_b.append(db)

    for i in range(len(weights) - 2, -1, -1):
        dA = np.dot(dZ, weights[i + 1].T)
        dZ = dA * activations_derivative[i](caches[i][1])
        dW = np.dot((X if i == 0 else caches[i - 1][0]).T, dZ) / m
        db = np.sum(dZ, axis=0, keepdims=True) / m
        grads_w.insert(0, dW)
        grads_b.insert(0, db)

    return grads_w, grads_b

# Plot accuracy
def plot_accuracies(accuracies, title):
    plt.figure(figsize=(10, 6))
    plt.plot(range(len(accuracies)), accuracies, marker='o')
    plt.title(title)
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.grid(True)
    plt.show()

# Evaluation
def evaluate_model(X, y, weights, biases, activations):
    A, _ = forward_propagation(X, weights, biases, activations)
    predictions = np.argmax(A, axis=1)
    accuracy = np.mean(predictions == y)
    print(f"Accuracy: {accuracy * 100:.2f}%")
    return accuracy

# Modified training loop with mini-batch gradient descent and Adam optimizer
def train_neural_network(X_train, y_train, layer_sizes, learning_rate, epochs, batch_size, hidden_activation, hidden_activation_derivative):
    weights, biases = initialize_weights(layer_sizes)
    y_train_encoded = one_hot_encode(y_train, layer_sizes[-1])
    accuracies = []
    n_batches = len(X_train) // batch_size

    # Use hidden_activation for all layers except the last, and always use softmax for the last layer
    activations = [hidden_activation] * (len(layer_sizes) - 2) + [softmax]
    activations_derivative = [hidden_activation_derivative] * (len(layer_sizes) - 2)

    # Adam optimizer parameters
    beta1, beta2 = 0.9, 0.999
    epsilon = 1e-8
    m_w = [np.zeros_like(w) for w in weights]
    v_w = [np.zeros_like(w) for w in weights]
    m_b = [np.zeros_like(b) for b in biases]
    v_b = [np.zeros_like(b) for b in biases]

    for epoch in range(epochs):
        epoch_loss = 0
        epoch_accuracy = 0

        # Shuffle the data
        permutation = np.random.permutation(len(X_train))
        X_train = X_train[permutation]
        y_train_encoded = y_train_encoded[permutation]

        for batch in range(n_batches):
            start = batch * batch_size
            end = start + batch_size
            X_batch = X_train[start:end]
            y_batch = y_train_encoded[start:end]

            A, caches = forward_propagation(X_batch, weights, biases, activations)
            loss = -np.mean(np.sum(y_batch * np.log(A + 1e-8), axis=1))
            grads_w, grads_b = back_propagation(X_batch, y_batch, weights, caches, activations_derivative)

            # Adam update
            for i in range(len(weights)):
                m_w[i] = beta1 * m_w[i] + (1 - beta1) * grads_w[i]
                v_w[i] = beta2 * v_w[i] + (1 - beta2) * (grads_w[i] ** 2)
                m_b[i] = beta1 * m_b[i] + (1 - beta1) * grads_b[i]
                v_b[i] = beta2 * v_b[i] + (1 - beta2) * (grads_b[i] ** 2)

                m_w_hat = m_w[i] / (1 - beta1 ** (epoch * n_batches + batch + 1))
                v_w_hat = v_w[i] / (1 - beta2 ** (epoch * n_batches + batch + 1))
                m_b_hat = m_b[i] / (1 - beta1 ** (epoch * n_batches + batch + 1))
                v_b_hat = v_b[i] / (1 - beta2 ** (epoch * n_batches + batch + 1))

                weights[i] -= learning_rate * m_w_hat / (np.sqrt(v_w_hat) + epsilon)
                biases[i] -= learning_rate * m_b_hat / (np.sqrt(v_b_hat) + epsilon)

            predictions = np.argmax(A, axis=1)
            accuracy = np.mean(predictions == np.argmax(y_batch, axis=1))
            
            epoch_loss += loss
            epoch_accuracy += accuracy

        epoch_loss /= n_batches
        epoch_accuracy /= n_batches
        accuracies.append(epoch_accuracy)

        if epoch % 10 == 0:
            print(f"Epoch {epoch}, Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy * 100:.2f}%")

    return weights, biases, accuracies

# Experimentation
if __name__ == "__main__":
    X_train, y_train, X_test, y_test = load_data()

    configurations = [
        ("1 Hidden Layer, 16 Neurons", [196, 16, 10]),
        ("1 Hidden Layer, 32 Neurons", [196, 32, 10]),
        ("1 Hidden Layer, 64 Neurons", [196, 64, 10]),
        ("2 Hidden Layers, 32 Neurons Each", [196, 32, 32, 10]),
        ("2 Hidden Layers, 64 Neurons Each", [196, 64, 64, 10]),
        ("2 Hidden Layers, 64 -> 32 Neurons", [196, 64, 32, 10]),
        ("2 Hidden Layers, 32 -> 64 Neurons", [196, 32, 64, 10]),
        ("3 Hidden Layers, 16 -> 16 -> 16", [196, 16, 16, 16, 10]),
        ("3 Hidden Layers, 32 -> 32 -> 32", [196, 32, 32, 32, 10]),
        ("3 Hidden Layers, 32 -> 16 -> 32", [196, 32, 16, 32, 10]),
        ("3 Hidden Layers, 32 -> 64 -> 32", [196, 32, 64, 32, 10]),
        ("3 Hidden Layers, 32 -> 64 -> 64", [196, 32, 64, 64, 10]),
    ]

    for name, layer_sizes in configurations:
        print(f"Running configuration: {name}")
        weights, biases, accuracies = train_neural_network(
            X_train, y_train, layer_sizes, 0.005, 50, 32, relu, relu_derivative
        )
        plot_accuracies(accuracies, name)
        evaluate_model(X_test, y_test, weights, biases, [relu] * (len(layer_sizes) - 2) + [softmax])

    # Repeat with tanh for 3-hidden layer networks
    tanh_configurations = [
        ("3 Hidden Layers-[Tanh], 16 -> 16 -> 16", [196, 16, 16, 16, 10]),
        ("3 Hidden Layers-[Tanh], 32 -> 32 -> 32", [196, 32, 32, 32, 10]),
        ("3 Hidden Layers-[Tanh], 32 -> 16 -> 32", [196, 32, 16, 32, 10]),
        ("3 Hidden Layers-[Tanh], 32 -> 64 -> 32", [196, 32, 64, 32, 10]),
        ("3 Hidden Layers-[Tanh], 32 -> 64 -> 64", [196, 32, 64, 64, 10]),
    ]

    for name, layer_sizes in tanh_configurations:
        print(f"Running configuration: {name}")
        weights, biases, accuracies = train_neural_network(
            X_train, y_train, layer_sizes, 0.005, 50, 32, tanh, tanh_derivative
        )
        plot_accuracies(accuracies, name)
        evaluate_model(X_test, y_test, weights, biases, [tanh] * (len(layer_sizes) - 2) + [softmax])
```

```python
import numpy as np

# Load the TinyMNIST dataset
def load_data():
    # Load the dataset files
    train_data = np.loadtxt('TinyMNIST/trainData.csv', delimiter=',')
    train_labels = np.loadtxt('TinyMNIST/trainLabels.csv', delimiter=',').astype(np.int64)
    test_data = np.loadtxt('TinyMNIST/testData.csv', delimiter=',')
    test_labels = np.loadtxt('TinyMNIST/testLabels.csv', delimiter=',').astype(np.int64)

    # Debug: Print shapes and data types
    print("Train data shape:", train_data.shape)
    print("Train labels shape:", train_labels.shape)
    print("Test data shape:", test_data.shape)
    print("Test labels shape:", test_labels.shape)
    print("Train data dtype:", train_data.dtype)
    print("Train labels dtype:", train_labels.dtype)

    # Normalize pixel values to [0, 1]
    X_train, X_test = train_data / 255.0, test_data / 255.0

    # Debug: Check normalization
    print("X_train min-max:", X_train.min(), X_train.max())
    print("X_test min-max:", X_test.min(), X_test.max())

    return X_train, train_labels, X_test, test_labels

# Activation functions
def relu(x):
    return np.maximum(0, x)

def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)

# Derivatives
def relu_derivative(x):
    return (x > 0).astype(float)

def one_hot_encode(y, num_classes):
    return np.eye(num_classes)[y]

# Initialize weights and biases
def initialize_weights(sizes):
    weights = [np.random.randn(sizes[i], sizes[i+1]) * np.sqrt(2.0/sizes[i]) for i in range(len(sizes)-1)]
    biases = [np.zeros((1, sizes[i+1])) for i in range(len(sizes)-1)]
    return weights, biases

# Forward propagation
def forward_propagation(X, weights, biases, activations):
    A = X
    caches = []
    for i in range(len(weights) - 1):
        Z = np.dot(A, weights[i]) + biases[i]
        A = activations[i](Z)
        caches.append((A, Z))
    Z = np.dot(A, weights[-1]) + biases[-1]
    A = softmax(Z)
    caches.append((A, Z))
    return A, caches

# Backpropagation
def back_propagation(X, y, weights, caches, activations_derivative):
    grads_w = []
    grads_b = []
    A, Z = caches[-1]
    m = X.shape[0]
    dZ = A - y
    dW = np.dot(caches[-2][0].T, dZ) / m
    db = np.sum(dZ, axis=0, keepdims=True) / m
    grads_w.append(dW)
    grads_b.append(db)

    for i in range(len(weights) - 2, -1, -1):
        dA = np.dot(dZ, weights[i + 1].T)
        dZ = dA * activations_derivative[i](caches[i][1])
        dW = np.dot((X if i == 0 else caches[i - 1][0]).T, dZ) / m
        db = np.sum(dZ, axis=0, keepdims=True) / m
        grads_w.insert(0, dW)
        grads_b.insert(0, db)

    return grads_w, grads_b

# Training loop with mini-batch gradient descent and Adam optimizer
def train_neural_network(X_train, y_train, layer_sizes, learning_rate, epochs, batch_size, activations, activations_derivative):
    weights, biases = initialize_weights(layer_sizes)
    y_train_encoded = one_hot_encode(y_train, layer_sizes[-1])
    accuracies = []
    n_batches = len(X_train) // batch_size

    # Adam optimizer parameters
    beta1, beta2 = 0.9, 0.999
    epsilon = 1e-8
    m_w = [np.zeros_like(w) for w in weights]
    v_w = [np.zeros_like(w) for w in weights]
    m_b = [np.zeros_like(b) for b in biases]
    v_b = [np.zeros_like(b) for b in biases]

    for epoch in range(epochs):
        epoch_loss = 0
        epoch_accuracy = 0

        # Shuffle the data
        permutation = np.random.permutation(len(X_train))
        X_train = X_train[permutation]
        y_train_encoded = y_train_encoded[permutation]

        for batch in range(n_batches):
            start = batch * batch_size
            end = start + batch_size
            X_batch = X_train[start:end]
            y_batch = y_train_encoded[start:end]

            A, caches = forward_propagation(X_batch, weights, biases, activations)
            loss = -np.mean(np.sum(y_batch * np.log(A + 1e-8), axis=1))
            grads_w, grads_b = back_propagation(X_batch, y_batch, weights, caches, activations_derivative)

            # Adam update
            for i in range(len(weights)):
                m_w[i] = beta1 * m_w[i] + (1 - beta1) * grads_w[i]
                v_w[i] = beta2 * v_w[i] + (1 - beta2) * (grads_w[i] ** 2)
                m_b[i] = beta1 * m_b[i] + (1 - beta1) * grads_b[i]
                v_b[i] = beta2 * v_b[i] + (1 - beta2) * (grads_b[i] ** 2)

                m_w_hat = m_w[i] / (1 - beta1 ** (epoch + 1))
                v_w_hat = v_w[i] / (1 - beta2 ** (epoch + 1))
                m_b_hat = m_b[i] / (1 - beta1 ** (epoch + 1))
                v_b_hat = v_b[i] / (1 - beta2 ** (epoch + 1))

                weights[i] -= learning_rate * m_w_hat / (np.sqrt(v_w_hat) + epsilon)
                biases[i] -= learning_rate * m_b_hat / (np.sqrt(v_b_hat) + epsilon)

            predictions = np.argmax(A, axis=1)
            accuracy = np.mean(predictions == np.argmax(y_batch, axis=1))
            
            epoch_loss += loss
            epoch_accuracy += accuracy

        epoch_loss /= n_batches
        epoch_accuracy /= n_batches
        accuracies.append(epoch_accuracy)

        if epoch % 10 == 0:
            print(f"Epoch {epoch}, Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy * 100:.2f}%")

    return weights, biases, accuracies

# Evaluation
def evaluate_model(X, y, weights, biases, activations):
    A, _ = forward_propagation(X, weights, biases, activations)
    predictions = np.argmax(A, axis=1)
    accuracy = np.mean(predictions == y)
    print(f"Accuracy: {accuracy * 100:.2f}%")
    return accuracy

# Debug: Check label distribution
def check_label_distribution(y):
    unique, counts = np.unique(y, return_counts=True)
    print("\nLabel distribution:")
    for label, count in zip(unique, counts):
        print(f"Label {label}: {count}")

# Main execution
if __name__ == "__main__":
    X_train, y_train, X_test, y_test = load_data()

    # Debug: Check label distribution
    print("Training set label distribution:")
    check_label_distribution(y_train)
    print("\nTest set label distribution:")
    check_label_distribution(y_test)

    configurations = [
        ("2 Hidden Layers, 128 -> 64 Neurons", [196, 128, 64, 10]),
    ]

    for name, layer_sizes in configurations:
        print(f"\nRunning configuration: {name}")
        weights, biases, accuracies = train_neural_network(
            X_train, y_train, layer_sizes, 0.001, 200, 32,
            [relu] * (len(layer_sizes) - 2) + [softmax],
            [relu_derivative] * (len(layer_sizes) - 2)
        )
        print("\nFinal accuracies:")
        print("Training set evaluation:")
        evaluate_model(X_train, y_train, weights, biases, [relu] * (len(layer_sizes) - 2) + [softmax])
        print("Test set evaluation:")
        evaluate_model(X_test, y_test, weights, biases, [relu] * (len(layer_sizes) - 2) + [softmax])

        # Debug: Check predictions
        A, _ = forward_propagation(X_test[:10], weights, biases, [relu] * (len(layer_sizes) - 2) + [softmax])
        predictions = np.argmax(A, axis=1)
        print("\nSample predictions vs actual:")
        for pred, actual in zip(predictions, y_test[:10]):
            print(f"Predicted: {pred}, Actual: {actual}")

    # Debug: Final model parameters
    print("\nFinal model parameters:")
    for i, (w, b) in enumerate(zip(weights, biases)):
        print(f"Layer {i+1}:")
        print(f"  Weights shape: {w.shape}, mean: {np.mean(w):.5f}, std: {np.std(w):.5f}")
        print(f"  Biases shape: {b.shape}, mean: {np.mean(b):.5f}, std: {np.std(b):.5f}")
```
